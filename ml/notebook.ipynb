```python
summarizer = pipeline('summarization', model='facebook/bart-large-cnn')
sample_text = "Курс по Machine Learning включает темы: регрессия, деревья решений, нейронные сети и многое другое."
print(summarizer(sample_text, max_length=40, min_length=10, do_sample=False)[0]['summary_text'])
```
